```python
def formatWellName1(name):
    temp = name
    temp = temp.replace("/", "_")
    temp = temp.replace("-", "_")
    
    space = 0
    for i in range(0,len(temp)):
        if(temp[i] == ' '):
            break
        space += 1
    if(space != len(temp)):
        temp = temp[0:space]
    
    return temp
```

```python
def formatWellName2(name):
    temp = name
    lastNum = temp.find("__")
            
    temp = temp[0:lastNum]
    
    return temp
```

```python
import numpy as np
import pandas as pd

from pandas import read_excel
```

```python
# find your sheet name at the bottom left of your excel file and assign 
# it to my_sheet 
file_name = 'public_core_images_excel_recording.xlsx' # name of your excel file
df = read_excel(file_name)
print(df.head())

unique = set([])

for i in range (0, len(df)):
    wellName = df["6406_3_2"][i]
    unique.add(formatWellName1(str(wellName)))

print(unique)

exportDF = pd.DataFrame(columns = ['WellName'])
temp = list(unique)
print(temp)
for i in range(0, len(temp)):
    exportDF = exportDF.append({'WellName': temp[i]}, ignore_index = True)

print(exportDF)
exportDF.to_csv("unique_public_core_images_excel_recording.csv")
```
